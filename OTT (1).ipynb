Converting the Video Streaming Platform(Y/N) columns into integer.

```python
def trans_con(x):
    if x =='Yes':
        return 1
    if x =='No':
        return 0
```

```python
data['Video Streaming Platform(Y/N)']=data['Video Streaming Platform(Y/N)'].apply(trans_con)
```

Converting the column(Frequency of VSPA) into integer.

```python
def trans_con(x):
    if x =='Daily':
        return 1
    if x =='Monthly':
        return 2
    if x =='Once or twice in a week':
        return 3
    if x =='None of the Above':
        return 0
```

```python
data['Frequency_of_VSPA']=data['Frequency_of_VSPA'].apply(trans_con)
```
