```python
GraphChart = data.sort_values(by='Overall Rating', ascending=False).sort_index()
GraphChart = GraphChart[['Overall Rating','Netflix Original Series','Prime Original Series','Hotstar Original Series','Hulu Original Series']]
GraphChart.style.bar(subset=["Netflix Original Series",], color='#c684e0')\
                 .bar(subset=["Prime Original Series"], color='#937bdb')\
                 .bar(subset=["Hotstar Original Series"], color='#5ae3ed')\
                 .bar(subset=["Hulu Original Series"], color='#73dec0')\
                 .bar(subset=["Overall Rating"],color='#').background_gradient(cmap='BrBG')
```
